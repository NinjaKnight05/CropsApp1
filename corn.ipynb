```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('market.csv')
```

```python
df
```

```python
df.rename(columns={"Unnamed: 0": "Commodity Group", "Unnamed: 1": "Commodity","Unnamed: 3":"Price 0","Unnamed: 2":"Quantity","Unnamed: 5":"Price 2","Unnamed: 6":"Arrived Material 1","Unnamed: 7":"Arrival Material 2","Unnamed: 8":"Arrival Material 3","Marketwise Price & Arrival Report (03-01-2026)":"Price 1"}, inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(0,inplace=True)
df.head()
```

```python
df.drop(1,inplace=True)
df.head()
```

```python
cols = ["Arrived Material 1", "Arrival Material 2", "Arrival Material 3"]

df[cols] = df[cols].replace('-', 0)
```

```python
df['Total Arrival']=df["Arrived Material 1"].astype(float)+df["Arrival Material 2"].astype(float)+df["Arrival Material 3"].astype(float)
df.head(3)
```

```python
cols1 = ["Price 0", "Price 1", "Price 2","Quantity"]

df[cols1] = df[cols1].replace('-', 0)
```

```python
df.reset_index(drop=True)
```

```python

px.pie(df,names='Commodity',values='Quantity',title='Crops & Quantity')
```

```python
px.bar(df,x='Commodity',y='Price 1',color='Quantity',title='Price of Crops with Mentioned Quantity <03-Jan-2026>')
```

```python
px.funnel(df,x='Commodity',y='Price 2',title='Price on 01-Jan-2026',color='Commodity Group')
```

```python
px.scatter(df,x='Commodity',y='Price 0',color='Commodity Group',title='Price on 02-Jan-2026 Same Quantity')
```

```python
px.pie(df,names='Commodity Group',values='Quantity',title='Seeds & Quantity')
```

```python
px.line(df,x='Commodity',y='Total Arrival')
```
